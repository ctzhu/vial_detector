```python
figure, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
for item, geno in zip(dfs, genotype):
    item = item[item.True_particle]
    _gpb = item.groupby('Timestamp')
    (-_gpb.y.mean()+item.y.max()).plot(ax=ax1)
    _gpb.y.count().plot(ax=ax2, label=geno)
ax1.set_ylabel('Vertial Distance')
ax1.set_xlabel('')
ax2.set_ylabel('Count of flies that moved')
plt.legend(loc=4, ncol=3)
```
