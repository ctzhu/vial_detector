#Simple analysis using `pandas`

```python
big_df = pd.concat(P_dfs)
big_df.head()
```

```python
big_df_clean = big_df[(big_df.y>5)&
                      big_df.True_particle]
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(6,9),
                               sharex=True, squeeze=True)
gby_obj = big_df_clean.groupby('Timestamp')
ax1.plot(gby_obj.count().index,
         -gby_obj.mean().y + big_df_clean.y.max())
ax1.set_title('Mean vertical displacement from top')
ax1.set_ylabel('pixel')

ax2.plot(gby_obj.count().index,
         gby_obj.count().x)
ax2.set_title('No. of flies detected')
ax2.set_ylabel('Count')
ax2.set_xlabel('Time (S)')
```
